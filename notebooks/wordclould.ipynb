```python
path = "../pdf-reports/plaintext/flying-2.0-enabling-automated-air-travel-by-identifying-and-addressing-the-challenges-of-iot-rfid-technology.json"
```

```python
import json

with open(path) as json_file:
    parsed = json.load(json_file)
```

```python
from collections import Counter

wordcount = Counter(parsed["content"].split())
wordcount.most_common(10)
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

wordcloud = WordCloud(background_color="white", max_words=1000)
wordcloud.generate_from_frequencies(wordcount)

plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```
