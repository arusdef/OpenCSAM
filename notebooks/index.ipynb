```python
import pandas as pd
from pathlib import Path
import json
import re

df = pd.DataFrame(columns=["filename", "title", "text"])

idx = 0
for path in Path("../pdf-reports/plaintext").iterdir():
    if path.suffix == '.json':
        with path.open() as json_file:
            parsed = json.load(json_file)
            filename = path.stem
            
            # Some documents do not have title among metadata.
            title = parsed["metadata"]["title"] if "title" in parsed["metadata"].keys() else filename
                        
            df.loc[idx] = [filename, title, parsed["content"]]
            idx += 1

df.head()
```

```python
from elasticsearch import Elasticsearch

es = Elasticsearch('http://ae00eaabc473b11e8b84c061e85c3662-188118179.eu-west-3.elb.amazonaws.com:9200/')

records = df.T.to_dict()
records = [records[i] for i in records]
for idx, row in enumerate(records):
    res = es.index(index="test-index", doc_type='pdf_documents', id=idx, body=row)
    print(res)
```
